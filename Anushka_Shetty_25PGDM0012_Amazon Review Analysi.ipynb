```python
import pandas as pd
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
from nrclex import NRCLex

# Load the dataset (No need to upload again)
file_name = "Reviews 500.csv"  # Ensure this file exists
df = pd.read_csv(file_name)

# Print column names to confirm
print("Column Names:", df.columns.tolist())

# Clean column names (remove spaces, make lowercase)
df.columns = df.columns.str.strip().str.lower()

# Use the "text" column for sentiment & emotion analysis
review_column = "text"

# Download NLTK data
nltk.download('vader_lexicon')

# Initialize sentiment analyzer
sia = SentimentIntensityAnalyzer()

# Function to analyze sentiment
def get_sentiment(text):
    if pd.isna(text):  # Handle missing values
        return "Neutral"

    scores = sia.polarity_scores(str(text))  # Convert to string to avoid errors
    if scores['compound'] >= 0.05:
        return "Positive"
    elif scores['compound'] <= -0.05:
        return "Negative"
    else:
        return "Neutral"

# Function to analyze emotions
def get_emotion(text):
    if pd.isna(text):
        return "No Emotion"

    emotion_scores = NRCLex(str(text)).top_emotions
    if emotion_scores:
        return emotion_scores[0][0]  # Return the most dominant emotion
    return "No Emotion"

# Apply sentiment & emotion analysis
df['Sentiment'] = df[review_column].apply(get_sentiment)
df['Emotion'] = df[review_column].apply(get_emotion)

# Save the result
output_file = "Reviews_with_Sentiment_Emotion.csv"
df.to_csv(output_file, index=False)

# Display sentiment & emotion distribution
print("\nSentiment Distribution:")
print(df['Sentiment'].value_counts())

print("\nTop Emotions Detected:")
print(df['Emotion'].value_counts())
```

```python
import pandas as pd
import nltk
from sklearn.feature_extraction.text import TfidfVectorizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter

# Load the dataset
file_name = "Reviews 500.csv"  # Ensure this file exists
df = pd.read_csv(file_name)

# Clean column names (remove spaces, make lowercase)
df.columns = df.columns.str.strip().str.lower()

# Use the "text" column for analysis
review_column = "text"

# Download NLTK data if not done already
nltk.download('stopwords')
nltk.download('punkt')

# Remove stopwords and tokenize text
stop_words = set(stopwords.words('english'))

def clean_and_tokenize(text):
    tokens = word_tokenize(str(text).lower())
    return [word for word in tokens if word.isalpha() and word not in stop_words]

# Apply tokenization and cleaning
df['cleaned_text'] = df[review_column].apply(clean_and_tokenize)

# **Word Frequency Analysis**
all_words = [word for sublist in df['cleaned_text'] for word in sublist]
word_counts = Counter(all_words)

# Most common words
most_common_words = word_counts.most_common(10)
print("\nMost Common Words:")
print(most_common_words)

# **Keyword Extraction using TF-IDF**
vectorizer = TfidfVectorizer(max_features=10)
X = vectorizer.fit_transform(df[review_column].dropna())
tfidf_words = vectorizer.get_feature_names_out()

print("\nTop 10 Keywords using TF-IDF:")
print(tfidf_words)

# **WordCloud Visualization**
wordcloud = WordCloud(width=800, height=400, background_color="white").generate_from_frequencies(word_counts)

# Display the wordcloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis('off')
plt.show()

# Save the word frequency and keyword analysis results
word_freq_file = "word_frequency_analysis.csv"
df_word_freq = pd.DataFrame(most_common_words, columns=["Word", "Frequency"])
df_word_freq.to_csv(word_freq_file, index=False)

print("\nWord frequency analysis saved to 'word_frequency_analysis.csv'")
```
